```python
#Create two subsets with unique and duplicate values. 
# create a subset of duplicate value 
Duplicate=CustData[CustData.Customer_ID.duplicated()]
print(Duplicate)
```

```python
# create a subset of Unique value 
Unique=CustData[CustData.Customer_ID.duplicated()==False]
print(Unique)
```

```python
#Create data set with list of customers whose customer value greater than 10000. 

CustValue = CustData[CustData['Customer_Value']>10000]
print(CustValue)
print("Total Count of Rows : ",CustValue.shape[0])
```

```python
#Create variables average revenue per trip and balance points in the customer data set. 

CustData=CustData.assign(average_revenue_per_trip=CustData.Customer_Value/CustData.Buy_Times,
                         Balance_Points=CustData.Points_Earned-CustData.Points_Redeemed)
print(CustData)
```
